# Parsing Dates (날짜 분석하기)

- landslide data set ( `산사태 데이터셋`) 

````python
# modules we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

# read in our data
landslides = pd.read_csv("../input/landslide-events/catalog.csv")

# set seed for reproducibility
np.random.seed(0)
````

````python
landslides.head()
````

````python
print(landslides['date'].head())
````

# Convert our date columns to datetime

날짜 열이 날짜로 인식되지 않는다는 것을 알게 되었으니, 이를 날짜로 인식할 수 있도록 변환하기.
<br>
이를 `"날짜 구문 분석(Parsing Date)"`이라고 합니다. 우리는 문자열을 받아들여 그 구성 요소를 식별할 수 있게하기.
<br>
<br>
가능한 날짜 부분은 많지만 가장 일반적인 것은 일자를 나타내는 `%d`, 월을 나타내는 `%m`, 두 자리 연도를 나타내는 `%y`, 네 자리 연도를 나타내는 `%Y` 등입니다.

```
1/17/07 has the format "%m/%d/%y"
17-1-2007 has the format "%d-%m-%Y"
```

````python
# create a new column, date_parsed, with the parsed dates
landslides['date_parsed'] = pd.to_datetime(landslides['date'], format="%m/%d/%y") #to_datetime 이용해서 format만바꿔주면된다.
````

````python
# print the first few rows
landslides['date_parsed'].head()
````

````python
# get the day of the month from the date_parsed column
day_of_month_landslides = landslides['date_parsed'].dt.day
day_of_month_landslides.head()
````

# Plot 

```
# remove na's
day_of_month_landslides = day_of_month_landslides.dropna()
```
