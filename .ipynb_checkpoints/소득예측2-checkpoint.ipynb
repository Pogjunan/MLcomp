```python
import numpy as np
import random
import os
from autogluon.tabular import TabularPredictor
import matplotlib.pyplot as plt
import seaborn as sns

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42) # Seed 고정
import pandas as pd

train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')

display(train.head(3))
# display(test.head(3))
```

```python
#train = train[train['Income'] != 0]
X = train.drop(['ID','Income'], axis=1)

y = train['Income']
```

```python
X.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 범주형과 정수형 특성 분리
categorical_features = []
numerical_features = []

for feature in X.columns[1:-1]:
    if X[feature].dtype == 'object':
        categorical_features.append(feature)
    else:
        numerical_features.append(feature)
sns.set(font_scale=2.0)
# 범주형 특성에 대한 barplot
plt.figure(figsize=(100, 200))
for idx, feature in enumerate(categorical_features):
    plt.subplot(10,2, idx + 1)
    plt.title(feature)
    sns.countplot(x=feature, data=train)

plt.tight_layout()
plt.show()
```

```python
categorical_features
```

```python

# 정수형 특성에 대한 histogram
plt.figure(figsize=(100, 600))
for idx, feature in enumerate(numerical_features):
    plt.subplot(7, 2, idx + 1)
    plt.title(feature)
    sns.histplot(x=feature, data=train, kde=True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train['Losses'] = np.where(X_train['Losses'] > 1000, 1000, X_train['Losses'])
```

```python
# 'Working_Week (Yearly)' 열의 값이 20보다 작은 경우를 0으로, 그렇지 않은 경우를 52로 대체
X_train['Working_Week (Yearly)'] = np.where(X_train['Working_Week (Yearly)'] < 20, 0, 52)
```

```python
from sklearn.preprocessing import LabelEncoder

encoding_target = list(X_train.dtypes[X_train.dtypes == "object"].index)

for i in encoding_target:
    le = LabelEncoder()
    
    # train과 test 데이터셋에서 해당 열의 모든 값을 문자열로 변환
    X_train[i] = X_train[i].astype(str)
    X_test[i] = X_test[i].astype(str)
    
    le.fit(X_train[i])
    X_train[i] = le.transform(X_train[i])
    
    # test 데이터의 새로운 카테고리에 대해 le.classes_ 배열에 추가
    for case in np.unique(X_test[i]):
        if case not in le.classes_: 
            le.classes_ = np.append(le.classes_, case)
    
    X_test[i] = le.transform(X_test[i])
```

```python
import numpy as np

def absolute_loss(predictions, y_test):
    """
    예측값과 실제값 사이의 절대값 손실을 계산하는 함수

    Parameters:
    predictions (array-like): 모델의 예측값 배열
    y_test (array-like): 실제값 배열

    Returns:
    float: 절대값 손실
    """
    # 절대값 손실 계산
    loss = np.mean(np.abs(predictions - y_test))
    return loss

# 예측값과 실제값을 가정하여 테스트
predictions = np.array([1.5, 2.5, 3.5, 4.5])
y_test = np.array([1, 2, 3, 4])

# 절대값 손실 계산
loss = absolute_loss(predictions, y_test)
print("Absolute Loss:", loss)
```
